```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load():
    global img
    original = cv2.imread("data/flat_chessboard.png")
    original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
    img = original.copy()
```

```python
def imshow(img):
    plt.figure(figsize=(7,7))
    plt.imshow(img, cmap = 'gray')
```

```python
load()
```

```python
imshow(img)
```

```python
# main thing here to keep in mind is that the image should have chessboard or grid like pattern.
```

```python
#                                             (8 rows, 8 columns)
found, corners = cv2.findChessboardCorners(img, (8,8))
found, corners
```

```python
# here we dont find anything for (8,8) because the last corner cannot be detected
```

```python
#found tells that whether corners are found or not
```

```python
#                                             (7 rows, 7 columns)
found, corners = cv2.findChessboardCorners(img, (7,7))
```

```python
# lets draw corners on the image
plotted = cv2.drawChessboardCorners(image=img, patternSize=(7,7), corners=corners, patternWasFound=found)
```

```python
imshow(plotted)
```

```python
# corners are shown by circle, and rows are shown by slant line
```

# finding circle grid

```python
def load():
    global img
    original = cv2.imread("data/dot_grid.png")
    original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
    img = original.copy()
```

```python
load()
```

```python
imshow(img)
```

```python
# number of dots in row, and number of dots in columns is 10, 10 respectively
```

```python
# findCirclesGrid(image, patternSize[, centers[, flags[, blobDetector]]])
# flag is not a required param,  see shift+tab to see what are flags and why used also the last line
found, corners = cv2.findCirclesGrid(image=img, patternSize=(10, 10),flags=cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
found
```

```python
detectedcorners = cv2.drawChessboardCorners(img, (10, 10), corners, found)
imshow(detectedcorners)
```
