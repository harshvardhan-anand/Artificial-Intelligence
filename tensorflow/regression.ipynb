```python
import numpy as np
```

```python
x_data=np.linspace(1,12, 50000)
```

```python
noise=np.random.randn(len(x_data))
```

```python
y_label=0.5*x_data+20+noise
```

```python
import pandas as pd
```

```python
xdf=pd.DataFrame(x_data, columns=['x'])
```

```python
ydf=pd.DataFrame(y_label, columns=['y'])
```

```python

my_data=pd.concat([xdf, ydf], axis=1)
```

```python
my_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
my_data.sample(100).plot(kind='scatter', x='x', y='y')
plt.show()
```

```python
import tensorflow as tf
```

```python
batch_size=10
```

```python
m=tf.Variable(0.0)
b=tf.Variable(0.0)
```

```python
xph=tf.placeholder(tf.float32, [batch_size])
yph=tf.placeholder(tf.float32, [batch_size])
```

```python
y_model=m*xph+b
```

```python
error=tf.reduce_sum(tf.square(yph-y_model))
```

```python
optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.001)
train=optimizer.minimize(error)
```

```python
init=tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    steps=1000
    sess.run(init)
    for index in range(steps):
        ind=np.random.randint(len(x_data), size=batch_size)
        feed={xph:x_data[ind], yph:y_label[ind]}
        sess.run(train, feed_dict=feed)
        #k=sess.run(m)
        #print(k)
        
    m_model, b_model=sess.run([m,b])
```

```python
y_pred=m_model*x_data+b_model
```

```python
my_data.sample(250).plot(kind='scatter', x='x', y='y')
plt.plot(x_data, y_pred, 'r')
plt.show()
```

# Estimator API

```python
feat_cols=[tf.feature_column.numeric_column('x', shape=[1])]
```

```python
model=tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainx, testx, trainy, testy=train_test_split(x_data, y_label,test_size=0.03)
```

```python
train_input_fn=tf.estimator.inputs.numpy_input_fn({'x':trainx}, trainy, batch_size=10, num_epochs=1000, shuffle=True)
```

```python
eval_input_fn=tf.estimator.inputs.numpy_input_fn({'x':testx}, testy, batch_size=2, num_epochs=1, shuffle=False)
```

```python
model.train(input_fn=train_input_fn)
```

```python
model.evaluate(input_fn=eval_input_fn, steps=1000)
```

```python
pred_fn=tf.estimator.inputs.numpy_input_fn({'x': np.linspace(0,3,1000)}, batch_size=4, shuffle=False)
```

```python
list(model.predict(input_fn=pred_fn))
```

```python
pridictions=[]
for i in model.predict(input_fn=pred_fn):
    pridictions.append(i['predictions'])
```
