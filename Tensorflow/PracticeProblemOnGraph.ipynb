```python
import tensorflow as tf
```

```python
a = tf.constant(5)
b = tf.constant(10)
```

```python
d = tf.add(a, b)
```

```python
c = tf.multiply(a, b)
```

```python
f = tf.add(d, c)
```

```python
e = tf.subtract(d, c)
```

```python
g = tf.divide(e, f)
```

```python
sess = tf.Session()
```

```python
sess.run(g)
```

```python
sess.close()
```
