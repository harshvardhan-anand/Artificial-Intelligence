```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(345)
N = 1000
x = np.random.rand(N)
```

```python
w_true = 5
b_true = 10
```

```python
y_true = (x * w_true) + b_true

plt.scatter(x, y_true)
```

```python
noise = np.random.normal(scale=0.1, size=[N])

y_true = (x * w_true) + b_true + noise

plt.scatter(x, y_true)
```

```python
import pandas as pd

pd.DataFrame({'x':x,'y_true':y_true}).corr()
```

```python
w = np.random.normal()
w
```

```python
b = np.random.normal()
b
```

```python
y_pred = lambda w, x, b: w * x + b
```

## Loss function and its derivative is taken from:
## https://ml-cheatsheet.readthedocs.io/en/latest/gradient_descent.html


## New weights


```python
def train(no_of_data_pt, x, y_true, w, b, learning_rate, epochs):

    for i in range(epochs):

        # minimum of loss function are calculated by finding the gradient(dE/dW, E is loss function)(gradient finded)
        error = y_true - y_pred(w, x, b)
        gradient_w = (1 / no_of_data_pt) * sum(((-2 * x) * (error)))
        gradient_b = (1 / no_of_data_pt) * sum(((-2) * (error)))

        # Updating old weight
        w = w - (gradient_w * learning_rate)
        b = b - (gradient_b * learning_rate)

        return w, b
```

```python
predict = lambda x, w, b: w * x + b
```

```python
w, b = train(no_of_data_pt=N,
             x=x,
             y_true=y_true,
             w=w,
             b=b,
             learning_rate=0.405,
             epochs=10)

w, b
```

```python
yp = predict(x,w,b)
```

```python
plt.scatter(x, y_true,color='red') #true data
plt.scatter(x,yp,color='blue')  # predicted data
```
